```python
import numpy as np
import pandas as pd


df = pd.read_csv(r"C:\Users\hp\Desktop\Jupyter_note_books_data_science\customer_churn_dataset.csv", encoding='latin1')
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df['SubscriptionType']=le.fit_transform(df['SubscriptionType'])
df['PaymentMethod']=le.fit_transform(df['PaymentMethod'])
df['ContractType']=le.fit_transform(df['ContractType'])
df['HasDependents']=le.fit_transform(df['HasDependents'])
df['Churn']=le.fit_transform(df['Churn'])


```

```python
df.info()
```

```python
df.head()
```

```python
x = df.drop(columns=['Churn'])
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split as tts

x_train,x_test,y_train,y_test = tts(x,y, test_size= 0.3, random_state= 47)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python

Random_model = RandomForestClassifier()
Random_model.fit(x_train, y_train)
```

```python
Logistic_Model = LogisticRegression()
Logistic_Model.fit(x_train, y_train)
```

```python
decision_model = DecisionTreeClassifier()
decision_model.fit(x_train, y_train)
```

```python
KN_model = KNeighborsClassifier()
KN_model.fit(x_train, y_train)
```

```python
Logistic_predictions = Logistic_Model.predict(x_test)
Logistic_predictions
```

```python
decision_predictions = decision_model.predict(x_test)
decision_predictions
```

```python
KN_predictions = KN_model.predict(x_test)
KN_predictions
```

```python
Random_predictions = Random_model.predict(x_test)
Random_predictions
```

```python
from sklearn.metrics import accuracy_score,r2_score,precision_score
# Accuracy Score
logistic_regression_accuracy_score = accuracy_score(y_test, Logistic_predictions)
decision_tree_classifier_accuracy_score = accuracy_score(y_test, decision_predictions)
KN_model_accuracy_score = accuracy_score(y_test, KN_predictions)
Random_model_accuracy_score = accuracy_score(y_test, Random_predictions)
# R2 Score
logistic_regression_R2_score = r2_score(y_test, Logistic_predictions)
decision_tree_classifier_R2_score = r2_score(y_test, decision_predictions)
KN_model_R2_score = r2_score(y_test, KN_predictions)
Random_model_R2_score = r2_score(y_test, Random_predictions)
# Precision Score
logistic_regression_Precision_score = precision_score(y_test, Logistic_predictions)
decision_tree_classifier_Precision_score = precision_score(y_test, decision_predictions)
KN_model_Precision_score = precision_score(y_test, KN_predictions)
Random_model_Precision_score = precision_score(y_test, Random_predictions)
# Printing
print("Accuracy_score   R2_score    Precision_score")
print("Logistic Regression: ", logistic_regression_accuracy_score, logistic_regression_R2_score,logistic_regression_Precision_score)
print("Decision Tree Classifier : ", decision_tree_classifier_accuracy_score, decision_tree_classifier_R2_score,decision_tree_classifier_Precision_score)
print("KN model  : ", KN_model_accuracy_score, KN_model_R2_score, KN_model_Precision_score)
print("Random Forest Model: ", Random_model_accuracy_score,Random_model_R2_score, Random_model_Precision_score)


```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=2,
    class_weight='balanced',
    random_state=42
)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
```

```python
import joblib

# Save your trained model
joblib.dump(model, "churn_model.pkl")


# Load it again later
model = joblib.load("churn_model.pkl")
```
